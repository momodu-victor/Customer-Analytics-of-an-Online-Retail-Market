### Data Integration

In order to build a user interface for the RFM analysis, we would need to connect the datasource to the visualization tool. Hence, there is a need for **"Data Integration"**

By connecting dashboard directly to the datasource, the data and visuals on the dashboard would update real time as data is ingested into the SQLite database.

### Requirements

* Downloaded Microsoft Power BI Desktop. I ensured to install 64 bit version because the Power BI Desktop is 64 bit version.


* Install SQLite ODBC driver. 


* An existing SQLite Database. Here, we used the retail db we created earlier in the project.

#### Importing SQLite Data to Power BI Using ODBC DSN

* Opened the correct version (x64) of ODBC after SQLite ODBC Driver was downloaded and installed.


* Click “Drivers” tab and make sure SQLite ODBC Driver(s) successfully  installed




* Click “User DSN” tab then click “Add”


* Select appropriate driver from the list, in my case it is “SQLite3 ODBC Driver”


* Click “Finish”


* Enter a name for the data source


* Click “Browse” and locate your SQLite database then click OK



Now you successfully created an ODBC DSN that can be used in Power BI to connect to SQLite.

#### Loading the Data into the Visualisation tool (Microsft PowerBI)

The following steps are used to load the data from the sqlite table into the visualisation tool.

* Open Power BI Desktop


* Select “ODBC” from “Get Data” then click “Connect”.


* Select the DSN you created earlier from the dropdown list then click OK.







* Click “Default or Custom” tab then click “Connect”







* Select the desired tables from the list then click load. This is we load the **rfm_seg table**  and **rfm_kmeans table** into PowerBI.






Then, the dynamic Customer Insight dashboard is created and ready to aid data-driven decisions. 

It contains the following:

* RFM Analysis


* Kmeans Clustering


* Action Plan
